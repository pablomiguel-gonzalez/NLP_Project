# **Text Cleaning**

```python
import re
import string

world_events.columns = world_events.columns.str.strip().str.replace(' ', '_').str.replace('/', '_')


# text cleaning function
def clean_text(text):
    text = text.lower()  # Lowercase
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)  # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

# Apply to the 'Event' column
world_events['Name_of_Incident_clean'] = world_events['Name_of_Incident'].apply(clean_text)
world_events['Impact_clean'] = world_events['Impact'].apply(clean_text)

print(world_events[['Name_of_Incident', 'Name_of_Incident_clean']].head())
```

```python
def convert_year(year):
    if isinstance(year, str):
        year = year.strip()
        if 'BC' in year:
            year_numeric = -int(re.sub('[^0-9]', '', year))
        else:
            year_numeric = int(re.sub('[^0-9]', '', year))
        return year_numeric
    else:
        return np.nan

world_events['Year_Clean'] = world_events['Year'].apply(convert_year)

# Century Column
def get_century(year):
    if pd.isnull(year):
        return np.nan
    if year < 0:
        return ((abs(year) - 1) // 100 + 1) * -1  # Negative centuries for BC
    else:
        return (year // 100 + 1)

world_events['Century'] = world_events['Year_Clean'].apply(get_century)
print(world_events[['Year', 'Year_Clean', 'Century']].head())
```

```python
# Light cleaning: lowercase and strip spaces
columns_to_normalize = ['Country', 'Place_Name']

for col in columns_to_normalize:
    world_events[col] = world_events[col].str.strip().str.lower()

# Quick check
print(world_events[['Country', 'Place_Name']].sample(5))
```

# **TD-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the cleaned event names
tfidf_matrix = tfidf_vectorizer.fit_transform(world_events['Name_of_Incident_clean'])

print(tfidf_matrix.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def search_event(query, top_n=3):
    # Clean query
    query_clean = clean_text(query)

    # Transform query TF-IDF vector
    query_vec = tfidf_vectorizer.transform([query_clean])

    # cosine similarity between the query and all events
    similarities = cosine_similarity(query_vec, tfidf_matrix).flatten()

    top_indices = similarities.argsort()[-top_n:][::-1]

    results = world_events.iloc[top_indices]
    return results[['Name_of_Incident', 'Impact', 'Year_Clean', 'Country', 'Place_Name']]
```

```python
# Try a sample search
search_event("battle in india")
```

```python
# 1st chatbot, only using even retrieval
def chat():
    print("Hello! I'm HistoryBot 🤖. Ask me about a historical event. Type 'exit' to leave.")

    while True:
        query = input("\nYou: ")
        if query.lower() == 'exit':
            print("HistoryBot: Goodbye! 👋")
            break

        results = search_event(query, top_n=1)

        if not results.empty:
            event_name = results.iloc[0]['Name_of_Incident']
            impact = results.iloc[0]['Impact']
            year = results.iloc[0]['Year_Clean']
            country = results.iloc[0]['Country']
            place = results.iloc[0]['Place_Name']

            # Format year nicely for BC
            if year < 0:
                year_str = f"{abs(year)} BC"
            else:
                year_str = str(year)

            response = f"📜 {event_name} ({year_str})\nImpact: {impact}\nLocation: {place.title()}, {country.title()}"
            print(f"HistoryBot: {response}")
        else:
            print("HistoryBot: Sorry, I couldn't find anything about that. Try another question.")

# Run chatbot
chat()
```

# **Synthetic Dataset for Intent Classification**

```python
# Synthetic data
synthetic_data = {
    'question': [
        # Date questions
        "When did the Battle of Hastings happen?",
        "What year was the fall of the Berlin Wall?",
        "When was Julius Caesar assassinated?",
        "In what year was the Declaration of Independence signed?",
        "What year did World War II start?",

        # Person questions
        "Who was the leader during the French Revolution?",
        "Tell me about Mahatma Gandhi.",
        "Who commanded the Allied forces on D-Day?",
        "Tell me about Cleopatra.",
        "Who invented the telephone?",

        # Event questions
        "Describe the fall of the Roman Empire.",
        "Tell me about the discovery of America.",
        "What happened during the Cold War?",
        "Explain the Industrial Revolution.",
        "Give me an important event in ancient Egypt.",

        # More (augment)
        "Who led the American Civil Rights Movement?",
        "When was the Great Fire of London?",
        "Tell me about the signing of the Magna Carta.",
        "Who was involved in the Boston Tea Party?",
        "Describe the Renaissance period.",
        "When was the Trojan War fought?",
        "Tell me about the Black Death.",
        "Who was Napoleon Bonaparte?",
        "What year did the Russian Revolution occur?",
        "Explain the Apollo 11 Moon landing."
    ],
    'intent': [
        # Labels match the questions
        "date", "date", "date", "date", "date",
        "person", "person", "person", "person", "person",
        "event", "event", "event", "event", "event",
        "person", "date", "event", "person", "event",
        "date", "event", "person", "date", "event"
    ]
}

# Create DataFrame
synthetic_df = pd.DataFrame(synthetic_data)

# Peek at the data
synthetic_df.sample(5)
```

# **Training Classifier, Logistic Regression**

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    synthetic_df['question'], synthetic_df['intent'], test_size=0.2, random_state=42
)

# Vectorize
intent_vectorizer = TfidfVectorizer()
X_train_vec = intent_vectorizer.fit_transform(X_train)
X_test_vec = intent_vectorizer.transform(X_test)

# Train model
intent_classifier = LogisticRegression(max_iter=1000)
intent_classifier.fit(X_train_vec, y_train)

# Predict and evaluate
y_pred = intent_classifier.predict(X_test_vec)
print(classification_report(y_test, y_pred))
```

# **Using Hyperparameter Tuning**

```python
## Lets use gridsearch to improve results
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

# Set up pipeline: Vectorizer + Classifier
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', LogisticRegression(max_iter=1000))
])

# Define the grid
param_grid = {
    'tfidf__ngram_range': [(1, 1), (1, 2), (1, 3)],  # unigrams, bigrams, trigrams
    'clf__C': [0.01, 0.1, 1.0, 5.0, 10.0, 50.0],    # wider C range
    'clf__class_weight': [None, 'balanced'],
    'clf__penalty': ['l2'],  # stick to L2 regularization
    'clf__solver': ['liblinear']  # good for small datasets
}


grid = GridSearchCV(pipeline, param_grid, cv=3, scoring='f1_macro', n_jobs=-1, verbose=2)

grid.fit(X_train, y_train)

# Best params
print("Best Parameters:\n", grid.best_params_)

# Evaluate on test set
best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)
print(classification_report(y_test, y_pred, zero_division=0))
```

# **ChatBot**

```python
## define chat again, with best model
def chat():
    print("Hello! I'm HistoryBot 🤖. Ask me about a historical event. Type 'exit' to leave.")

    while True:
        query = input("\nYou: ")
        if query.lower() == 'exit':
            print("HistoryBot: Goodbye! 👋")
            break

        # Step 1: Predict Intent
        query_clean = clean_text(query)
        query_vec = intent_vectorizer.transform([query_clean])
        intent = best_model.predict(query_vec)[0]

        # Step 2: Search for Event
        results = search_event(query, top_n=1)

        if not results.empty:
            event_name = results.iloc[0]['Name_of_Incident']
            impact = results.iloc[0]['Impact']
            year = results.iloc[0]['Year_Clean']
            country = results.iloc[0]['Country']
            place = results.iloc[0]['Place_Name']

            if year < 0:
                year_str = f"{abs(year)} BC"
            else:
                year_str = str(year)

            # Step 3: Respond Based on Intent
            if intent == "date":
                response = f"🗓️ The event '{event_name}' happened in {year_str}."
            elif intent == "person":
                response = f"🧑‍🎓 '{event_name}' involved notable figures. Would you like to know more about the key people involved?"
            else:  # event
                response = f"📜 {event_name} ({year_str})\nImpact: {impact}\nLocation: {place.title()}, {country.title()}"

            print(f"HistoryBot: {response}")
        else:
            print("HistoryBot: Sorry, I couldn't find anything about that. Try another question.")
```
